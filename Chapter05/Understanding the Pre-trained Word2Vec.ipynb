# Understanding the Pre-trained Word2Vec

### Importing the libraries and modules

```python
import gensim
from gensim.models import KeyedVectors
```

### Load pre-trained vectors

```python
model = KeyedVectors.load_word2vec_format('/Users/amankedia/Desktop/Sunday/nlp-book/Chapter 5/Code/GoogleNews-vectors-negative300.bin', binary=True)
```

### Vocabulary size

```python
len(model.wv.vocab)
```

### Word vector size

```python
model.vector_size
```

### Vocabulary

```python
model.wv.vocab
```

### Word vector for India

```python
model.wv['India']
```

### Finding the word most similar to Delhi

```python
model.most_similar('Delhi')
```

### Validating the Relationship : vector(Man)  vector (King) + vector (Queen) = vector (Woman)

```python
result = model.most_similar(positive=['man', 'queen'], negative=['king'], topn=1)
print(result)
```

```python
result = model.most_similar(positive=['man', 'queen'], negative=['king'], topn=2)
print(result)
```

### Validating the Relationship : vector(France) + vector(Rome) - vector(Italy) = vector (Paris)

```python
result = model.most_similar(positive=['France', 'Rome'], negative=['Italy'], topn=1)
print(result)
```
